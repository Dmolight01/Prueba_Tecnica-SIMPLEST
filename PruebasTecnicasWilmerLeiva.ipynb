```python
# Codigo Realizado por Wilmer Mario Leiva Esteban
# Importacion de las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Encabezado de la presentación
print("📊Reto Tecnico – SIMPLEST Guru: Análisis básico de datos con Python. \nJunior AI Specialist.")

# Creacion conjunto de la lista de números
ventas = [5, 8, 2, 10, 3]
print("\nLista original:", ventas)

# Calculos metricas estadisticas: media y desviación estándar
media = np.mean(ventas)
desviacion_estandar = np.std(ventas)

print("\nMedia:", media)
print("Desviación estándar:", desviacion_estandar)

# Convertir la lista en un DataFrame de pandas
df = pd.DataFrame({'Ventas': ventas})
print("\nDataFrame:")
print(df)

# (Bonus): Clasificar las ventas como 'alta' o 'baja'
df['Nivel'] = df['Ventas'].apply(lambda x: 'alta' if x > 6 else 'baja')
print("\nContenido del DataFrame:")
print(df)

# Graficar los valores usando matplotlib

# Colores: verde para 'alta', rojo para 'baja'
colors = df['Nivel'].map({'alta': 'green', 'baja': 'red'})
plt.bar(df.index, df['Ventas'], color=colors)
plt.title('Simulación de Ventas Clasificadas por Nivel')
plt.xlabel('Índice')
plt.ylabel('Valor de la Venta')
plt.xticks(df.index)

# Agregar texto encima de cada barra
for i, (valor, nivel) in enumerate(zip(df['Ventas'], df['Nivel'])):
    plt.text(i, valor + 0.3, nivel, ha='center', fontsize=10)

# Mejorar presentación
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# Codigo Realizado por Wilmer Mario Leiva Esteban
```
